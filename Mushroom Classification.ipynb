# Setting the Data for Modelling

```python
X = df.drop('class', axis=1)
```

```python
X
```

As features are categorical letters, converting to dummy variables

```python
X = pd.get_dummies(X, drop_first = True)
```

```python
X
```

```python
y = df['class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
#splitting data for training set and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=101)
```

```python
print(X_train.shape,
X_test.shape,
y_train.shape,
y_test.shape)
```

```python
#import AdaBoostClassifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
model = AdaBoostClassifier(n_estimators=1)
```

Here we are giving n_estimators = 1 (i.e., the AdaboostClassifier has to find one single best feature to split on.)

```python
model.fit(X_train, y_train)
```

# EVALUATION

```python
from sklearn.metrics import classification_report, plot_confusion_matrix, confusion_matrix, accuracy_score
```

```python
preds = model.predict(X_test)
```

```python
print(classification_report(y_test, preds))
```

Figuring out What it's actually splitting on

```python
model.feature_importances_
```

Here I can see that, it returned bunch of zeroes and only '1'.

That 1 feature is very important, because splitting is happening from that feature!!

```python
#Printing that 1 feature
model.feature_importances_.argmax()
```

It says that index 22 is the most important feature, that this model is using

```python
X.columns[22]
```

It says 'odor_n', According to our model, the best way to figure out whether a mushroom is poisonous or edible is by checking whether the mushroom has odor or not! and this model predicted with 88% accuracy

Now Can we improve this model....?

Yes!! by adding more stumps

```python
sns.countplot(data=df, x='odor', hue='class')
```

In the Above Visualization  we can see that the mushrooms which has no 'odor' are most likely 'EDIBLE'

So we can say that if you want to know a mushroom is poisonous or edible, ITS SIMPLE you can just know it just by smelling it. If it has no odor it is EDIBLE, if it has any odor then it is POISONOUS.

ITS not 100% because there are some more(a=almond, l=anise)that has odor but they are EDIBLE!! and they are some POISONOUS mushrooms that have no 'odor'

```python
error_rates = []

for n in range(1,96):
    model = AdaBoostClassifier(n_estimators=n)
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    
    error = 1-accuracy_score(y_test,predictions)
    error_rates.append(error)
```

```python
plt.plot(error_rates)
```

Previously I have choosen n_estimators = 1 and it performed well with 88% accuracy...
Now what are the number of estimators we can actually use??


It can be seen that in above Visualization as the  number of estimators are increasing, the error rate is decreasing Obviously model accuracy is increased. At certain number of estimators i.e, n_estimators = 20 there is a limit, Where if you increase no. of estimators beyond 20 there is no increase in model performance.
